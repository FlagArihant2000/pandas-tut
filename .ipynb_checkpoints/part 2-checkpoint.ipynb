```python
import pandas as pd
import matplotlib.pyplot as plt

print(pd.__version__)
```

```python
df = pd.read_csv("datasets/avocado.csv")
df["Date"] = pd.to_datetime(df["Date"])
```

```python
albany_df = df[df['region']=="Albany"]
albany_df.set_index("Date", inplace=True)
```

```python
albany_df["AveragePrice"].plot()
```

```python
# The graph is quite a mess, which can be made more visually appealing by smoothing it using moving averages

albany_df["AveragePrice"].rolling(25).mean().plot()
```

```python
# In case the graph is still ugly, we shall now look at the data,
albany_df.index # Note that the dates are not arranged properly as per year and month with day.
```

```python
albany_df.sort_index(inplace = True) # proper sorting of the date indexes
```

```python
albany_df["AveragePrice"].rolling(25).mean().plot()
```

```python
# Now we have a good graph
# let us save all this,
albany_df["price25ma"] = albany_df["AveragePrice"].rolling(25).mean()
```

```python
albany_df.tail()
```

```python
graph_df = pd.DataFrame()

for region in df['region'].unique()[:5]:
    print(region)
    region_df = df.copy()[df['region']==region]
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}_price25ma"] = region_df["AveragePrice"].rolling(25).mean()

    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]]  # note the double square brackets!
    else:
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])
```

```python
graph_df.tail()
```

```python
graph_df.plot(figsize=(8,5), legend=False)
```
